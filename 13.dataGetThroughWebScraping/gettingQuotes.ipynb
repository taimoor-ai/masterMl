```python
header={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://quotes.toscrape.com/page/1/', headers=header).text
soup=BeautifulSoup(webpage, 'html.parser')  
```

```python

finalDf=pd.DataFrame(columns=['text', 'author', 'tags'])
for i in range(1, 11):
    webpage=requests.get(f'https://quotes.toscrape.com/page/{i}/', headers=header).text
    soup=BeautifulSoup(webpage, 'html.parser')  
    quotes=soup.find_all('div', class_='quote')
    text=[]
    author=[]
    tags=[]
    for quote in quotes:
      text.append(quote.find('span', class_='text').text)
      author.append(quote.find('small', class_='author').text)
      tags.append(quote.find('meta', class_='keywords')['content'])
    tempDf=pd.DataFrame({'text':text, 'author':author, 'tags':tags})
    finalDf=pd.concat([finalDf, tempDf], ignore_index=True)
```
