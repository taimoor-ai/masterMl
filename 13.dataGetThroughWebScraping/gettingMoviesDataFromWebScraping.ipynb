```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.imdb.com/chart/top/?ref_=hm_nv_menu&view=grid', headers=headers).text
soup=BeautifulSoup(webpage, 'html.parser')
```

```python
MovieHeading=soup.find_all('h1', class_='ipc-title__text')
MovieHeading
```

```python
movies= soup.find_all('div', class_='ipc-poster-card')
movies[0]
```

```python
name=[]
rating=[]
release_year=[]
director=[]
for movie in movies:
    name.append(movie.find('a', class_='ipc-poster-card__title').text)
    rating.append(movie.find('span', class_='ipc-rating-star--rating').text)
    release_year.append(movie.find('div', class_='sc-ed99f764-4').text)
```

```python
print(len(name))
print(len(rating))
print(len(release_year))

movies={'name':name, 'rating':rating, 'release_year':release_year}
movies_df=pd.DataFrame(movies, columns=['name', 'rating', 'release_year'])

movies_df.to_json('movies.json', orient='records')
```
