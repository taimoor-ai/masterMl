```python
from sklearn.model_selection import train_test_split


X_train,X_test,Y_train,Y_test=   train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
from sklearn.impute   import SimpleImputer


se=SimpleImputer()

X_train_fever=  se.fit_transform(X_train[['fever']])

X_test_fever=  se.fit_transform(X_test[['fever']])


X_test_fever.shape
```

```python
#ordinal encoding for gender and cough 
from  sklearn.preprocessing   import OrdinalEncoder


OE=  OrdinalEncoder(categories=[['Mild','Strong']])


X_train_Cough=   OE.fit_transform(X_train[['cough']])

X_test_Cough=   OE.fit_transform(X_test[['cough']])


X_train_Cough.shape
```

```python
## one Hot Encoding on the gender and city

from sklearn.preprocessing  import OneHotEncoder



OHE=  OneHotEncoder(drop='first')


X_train_OHE=  OHE.fit_transform(X_train[['gender','city']]).value


X_test_OHE   = OHE.fit_transform(X_test[['gender','city']])


X_train_OHE.shape

```
